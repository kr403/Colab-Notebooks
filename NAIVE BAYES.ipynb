```python
import pandas as pd
import numpy as np
```

```python
#Upload dataset
from google.colab import files
uploaded = files.upload()
```

```python
#Read dataset
import io
ttt = pd.read_csv(io.BytesIO(uploaded['tic-tac-toe.csv']))
ttt = ttt.replace({
    #Covert output values to numerical values from categorical values
    "Class":{
        "negative":0,
        "positive":1
    }
})
#Spliting the features and the label
ttt_x = ttt.loc[:, ttt.columns != "Class"]
ttt_y = ttt.loc[:, "Class"]
```

#Without stratification

```python
#Split dataset into 90% traning data and 10% testing data
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(ttt_x, ttt_y, train_size=0.9)
train_y = list(train_y)
#P(y) when y = 0
P_0 = train_y.count(0)/len(train_y)
#P(y) when y = 1
P_1 = train_y.count(1)/len(train_y)
```

Fit function implementation(Without stratification)

```python
#Initializing a blank probablity dictonary
probability = {}

def fit(feature , label):
    #Find out the unique values in every feature
    unique_value = train_x[feature].unique()
    #Zipping the feature and label as a list
    merge = list(zip(train_x[feature], label))
    #If feature is not available in probablity dictonary 
    #then add it to the dictonary as a key and the value will be blank
    if(feature not in probability):
        probability[feature] = {}
    for a in unique_value:
        #x,o,b when y = 0
        t1 = (a, 0)
        #x,o,b when y = 1
        t2 = (a, 1)
        #y = 0
        value_0 = calculate_probability(t1, merge)
        #y = 1
        value_1 = calculate_probability(t2, merge)
        #Check x/o/b is available in probablity dictonary under feature key
        #if not then add it
        if a not in probability[feature]:
            probability[feature][a] = {}
            probability[feature][a]['0'] = value_0
            probability[feature][a]['1'] = value_1

def calculate_probability(t1, merge):
    #Count the total occurrence of x/o/b in the merge list
    p_xy = merge.count(t1)
    #Calculating the probablity for x/o/b
    probability = p_xy/train_y.count(t1[1])
    return probability
#Iterate by features
for a in train_x:
    fit(a, train_y)
print(probability)
```

Predict function implementation(Without stratification)

```python
#Initializing blank predict_result list
predict_result = []
def predict(feature):
    p_0 = 1
    p_1 = 1
    i = 0
    #Calculate probablity
    for a in feature:
        p_0 *= probability[feature.index[i]][a]['0']
        p_1 *= probability[feature.index[i]][a]['1']
        i += 1
    probability_no = p_0/(p_0 + p_1)
    probability_yes = p_1/(p_0 + p_1)
    if probability_no > probability_yes:
        predict_result.append(0)
    else:
        predict_result.append(1)
    return predict_result
#Iterate by rows
for a in test_x.iterrows():
    predict_result = predict(a[1])
print(predict_result)
```

Determining the accuracy, precision, recall and F1 score(Without stratification)

```python
test_y = list(test_y)
count = 0
i = 0
for a in test_y:
    if a == predict_result[i]:
        count += 1
    i += 1
accuracy = count * 100/ len(test_y)
print("The accuracy of this algorithm is:{}%".format(accuracy))
tp = (1, 1)
tn = (1, 0)
fp = (0, 1)
fn = (0, 0)
mer = list(zip(predict_result, test_y))
true_positive = mer.count(tp)
true_negative = mer.count(tn)
false_positive = mer.count(fp)
false_negative = mer.count(fn)
precission = true_positive/(true_positive + false_positive)
recall = true_positive/(true_positive + false_negative)
f1_score = (2*precission*recall)/precission + recall
print("The precision value of this algorithm is: {}".format(precission))
print("The recall value of this algorithm is: {}".format(recall))
print("F1 score of this algorithm is: {}".format(f1_score))
```
