```python
my_list = [5,10,5,30,25,10,10]
num_of_rounds = int(input())
for i in range(num_of_rounds):
    index_1 = int(input())
    index_2 = int(input())
    if index_1 >= 0 and index_1 < len(my_list):
        if index_2 >= 0 and index_2 < len(my_list):
            x = my_list[index_1]
            my_list[index_1] = my_list[index_2]
            my_list[index_2] = x
            print(my_list)
    else:
        print("Try again")
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(99)

trails = 10000000
hits = 0
hit_x = []
hit_y = []
miss_x = []
miss_y = []

for i in range(trails):
    x = np.random.uniform(low = -2, high = 5)
    y = np.random.uniform(low = 1, high = 7)
    if(x-2)**2 +(y-4)**2 <= 2**2:
        if(y >=4):
            hits += 1
            hit_x.append(x)
            hit_y.append(y)
        else:
            miss_x.append(x)
            miss_y.append(y)
    else:
        miss_x.append(x)
        miss_y.append(y)
est_pi = ((42)/2)*(hits/trails)
print(est_pi)
plt.scatter(hit_x,hit_y, c='red')
plt.scatter(miss_x, miss_y, c='blue')
plt.show()
```
