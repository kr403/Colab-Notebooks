# Read Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data =pd.read_csv('Titanic_train.csv')
data.head()
```

```python
for col in data.columns:
    print(col, data[col].nunique(), data[col].unique())
```

# Check Null values

```python
print(data.shape,'\n')
print(data.isnull().sum(),'\n')   # Null values in a individual column

sns.heatmap(data.isna())          # Heatmap where we can see the null values
```

```python
# Data Info: Column, # of non_null values & Data types
data.info()
```

```python
# Data Description
data.describe()
```

# Handle Null value for "Age" column

```python
print( (data['Age'].isna().sum()/len(data['Age']))*100 )   #  % of null values

sns.displot(x='Age',data=data)                             #  data distribution for "Age" column
```

```python
# Fill Null values by Age mean
print(data['Age'].isna().sum())                       # Null values before 
data['Age'].fillna(data['Age'].mean(),inplace=True)
print(data['Age'].isna().sum())                       # Null values after

sns.heatmap(data.isna())
```

# Handle Null value for "Cabin" column

```python
print( (data['Cabin'].isna().sum()/len(data['Cabin']))*100 )   #  % of null values

data.drop('Cabin',axis=1,inplace=True)                         #  Drop Cabin Column (Axis =1 means whole column)
data.head()
```

# Handle Null value for "Embarked" column 

```python
print( (data['Embarked'].isna().sum()/len(data['Embarked']))*100)   #  % of null values

print(data.shape)
print(data['Embarked'].isnull().sum())
data = data.dropna()
print(data.shape)
print(data['Embarked'].isnull().sum())


```

# countplot of subrvived vs not  survived
Survived=1, Not Survived=0

```python
sns.countplot(x='Survived',data=data)
```

# Male vs Female Survived

```python
sns.countplot(x='Survived',data=data,hue='Sex')
```

# Drop Unnecessary Columns

```python
data.head()
```

```python
data.drop(['PassengerId','Name','Ticket','Embarked'],axis=1,inplace=True)
data.head()
```

# Categorical column "Sex" Handle

```python
print('Sex', data['Sex'].nunique(), data['Sex'].unique())
data['Sex'].replace(['male', 'female'],[1, 0], inplace=True)
print('Sex', data['Sex'].nunique(), data['Sex'].unique())
```

# Train_Test Split

```python
from sklearn.model_selection import train_test_split

data_x = data.loc[:, data.columns != 'Survived']
data_y = data['Survived']

seed =1
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, 
                                                    train_size=0.85, 
                                                    random_state = seed)

print(train_x.shape,test_x.shape)
```

# Building Model using Logestic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()

# Fit
lr.fit(train_x, train_y) 

# Predict
predict = lr.predict(test_x)
```

# See Performance by confusion Matrix

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(test_y,predict),columns=['Actual Yes','Actual No'],index=['Predicted Yes','Predicted No'])
```

# Performance of Logistic regression


```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score 
   
acc = accuracy_score(test_y, predict)
print('Accuracy using Logistic regression = ',acc)

f1 = f1_score(test_y,predict)
print('F1_score using Logistic regression = ',f1)
                
precision = precision_score(test_y,predict)
print('Precision Score using Logistic regression = ',precision)

recall = recall_score(test_y,predict)
print('Recall score using Logistic regression = ',recall)
```

# Building Model using Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()

# input layer = 6
# hidden layer = 12
# hidden layer = 8
# hidden layer = 8
# output layer = 1

hl1 = Dense(12, input_dim=6, activation='relu')
hl2 = Dense(8, activation='relu')
hl3 = Dense(8, activation='relu')
ol = Dense(1, activation='sigmoid')
model.add(hl1)
model.add(hl2)
model.add(hl3)
model.add(ol)

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Fit the Neural Network Model

```python
model.fit(train_x, train_y, epochs=100, batch_size=16)
```

## Predict Survival's

```python
predict_y = model.predict(test_x)
py = []
for y in predict_y:
    if y>0.5: py.append(1)
    else: py.append(0)
print(py)
```

## Visualize the Neural Network

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model.png', show_shapes=True, show_layer_names=True)
```

## See Performance by confusion Matrix

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(test_y,py),columns=['Actual Yes','Actual No'],index=['Predicted Yes','Predicted No'])
```

## Performance of Neural Network

*   List item
*   List item



```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score 
   
acc = accuracy_score(test_y, py)
print('Accuracy using Neural Network = ',acc)

f1 = f1_score(test_y,py)
print('F1_score using Neural Network = ',f1)
                
precision = precision_score(test_y,py)
print('Precision Score using Neural Network = ',precision)

recall = recall_score(test_y,py)
print('Recall score using Neural Network = ',recall)
```
