```python
T_x = [0]
T_y = [10]
Tc_x = [20]
Tc_y = [10]

for t in range(0, 20):
    x = T_x[t]+7
    y = 10
    T_x.append(x)
    T_y.append(y)

for t in range(0, 20):
    x = 20
    y = Tc_y[t]-3
    Tc_x.append(x)
    Tc_y.append(y)
    
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
plt.plot(T_x, T_y, c='red', label='A')
plt.scatter(T_x, T_y, c='red')
plt.plot(Tc_x, Tc_y, c='blue', label='B')
plt.scatter(Tc_x, Tc_y, c='blue')
plt.legend()
plt.show()

import math 

P_x = [40]
P_y = [0]
S = 2

for t in range(0, 20):
    if t<10:
        dist = math.sqrt((T_x[t]-P_x[t])**2+(T_y[t]-P_y[t])**2)
        print("Distance between A to C is ",dist)
        cos_theta = (T_x[t]-P_x[t])/dist
        sin_theta = (T_y[t]-P_y[t])/dist

        new_px = P_x[t] + S*cos_theta
        new_py = P_y[t] + S*sin_theta
        
        P_x.append(new_px)
        P_y.append(new_py)
    else:
        dist = math.sqrt((Tc_x[t]-P_x[t])**2+(Tc_y[t]-P_y[t])**2)
        print("Distance between B to C is ",dist)
        cos_theta = (Tc_x[t]-P_x[t])/dist
        sin_theta = (Tc_y[t]-P_y[t])/dist

        new_px = P_x[t] + S*cos_theta
        new_py = P_y[t] + S*sin_theta
        
        P_x.append(new_px)
        P_y.append(new_py)

plt.figure(figsize=(8, 8))
plt.plot(T_x, T_y, c='red', label='A')
plt.scatter(T_x, T_y, c='red')
plt.plot(Tc_x, Tc_y, c='blue', label='B')
plt.scatter(Tc_x, Tc_y, c='blue')
plt.plot(P_x, P_y, c='green', label='C')
plt.scatter(P_x, P_y, c='green')
plt.legend()
plt.show()
```
