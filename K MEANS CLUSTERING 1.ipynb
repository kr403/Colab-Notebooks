```python
import copy
iteration = 0
while True:
    Temp_Clusters = [[],[],[],[],[],[]]
    cout=0
    for row in Data:
        X=row
        i=0
        cin=0
        mindistance=0
        for row in Centers:
            Y=row
            distance = ((((X[0] - Y[0] )**2) + ((X[1]-Y[1])**2) )**0.5)
            if cin==0:
                mindistance=distance
            else:
                if distance < mindistance:
                    mindistance=distance
                    i=cin
            cin=cin+1
        tc=Temp_Clusters[i]
        tc.append(cout)
        cout=cout+1
    cout=0
    for row in Temp_Clusters:
        s0=0
        s1=0
        avg0=0
        avg1=0
        l=len(row)
        for row in row:
            s0=s0+Data[row][0]
            s1=s1+Data[row][1]
        avg0=s0/l
        avg1=s1/l
        Centers[cout][0]=avg0
        Centers[cout][1]=avg1
        cout=cout+1
    iteration=iteration+1


    if iteration>1:
        shift=0
        cout=0
        for row in Data:
            cout0=0
            for a in Clusters:
                cout1=0
                for b in a:
                    if b==cout:
                        q=0
                        for p in Temp_Clusters[cout0]:
                            if(p==b):
                                q=1
                        if(q==0):
                            shift=shift+1
                    cout1=cout1+1
                cout0=cout0+1
            cout=cout+1
        print(shift)
        if shift<10:
            Clusters=Temp_Clusters
            break
    Clusters=Temp_Clusters
```
