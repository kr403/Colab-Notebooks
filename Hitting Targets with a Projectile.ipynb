```python
import math
import numpy as np
import matplotlib.pyplot as plt

Ta_x = []
Ta_y = []

x = 10
y = 10

# for 20 time units
for t in range(0, 21):
    x = (np.pi)*t
    y = t**2
    Ta_x.append(x)
    Ta_y.append(y)
    

plt.figure(figsize=(8, 8))
plt.plot(Ta_x, Ta_y, marker='o', label='A')
plt.legend()
plt.show()
 

Pb_x = [0]
Pb_y = [400]
Pc_x = [100]
Pc_y = [400]
Pd_x = [100]
Pd_y = [0]

Sb = 15
Sc = 20
Sd = 10
destroy_dist = 5

for t in range(0, 20):
    dist = math.sqrt((Ta_x[t]-Pb_x[t])**2+(Ta_y[t]-Pb_y[t])**2)
    # Destroy condition check
    if dist < destroy_dist:
        print("Target destroyed ->A!!! Time: ", t)
        break 

    cos_theta = (Ta_x[t]-Pb_x[t])/dist
    sin_theta = (Ta_y[t]-Pb_y[t])/dist

    new_pbx = Pb_x[t] + Sb*cos_theta
    new_pby = Pb_y[t] + Sb*sin_theta
    
    Pb_x.append(new_pbx)
    Pb_y.append(new_pby)

    dist = math.sqrt((Pb_x[t]-Pc_x[t])**2+(Pb_y[t]-Pc_y[t])**2)
    # Destroy condition check
    if dist < destroy_dist:
        print("Target destroyed ->B!!! Time: ", t)
        break 

    cos_theta = (Pb_x[t]-Pc_x[t])/dist
    sin_theta = (Pb_y[t]-Pc_y[t])/dist

    new_pcx = Pc_x[t] + Sc*cos_theta
    new_pcy = Pc_y[t] + Sc*sin_theta
    
    Pc_x.append(new_pcx)
    Pc_y.append(new_pcy)

    dist = math.sqrt((Pc_x[t]-Pd_x[t])**2+(Pc_y[t]-Pd_y[t])**2)
    # Destroy condition check
    if dist < destroy_dist:
        print("Target destroyed ->C!!! Time: ", t)
        break 

    cos_theta = (Pc_x[t]-Pd_x[t])/dist
    sin_theta = (Pc_y[t]-Pd_y[t])/dist

    new_pdx = Pd_x[t] + Sd*cos_theta
    new_pdy = Pd_y[t] + Sd*sin_theta
    
    Pd_x.append(new_pdx)
    Pd_y.append(new_pdy)


plt.figure(figsize=(8, 8))
plt.plot(Ta_x, Ta_y, marker='o', label='A')
plt.plot(Pb_x, Pb_y, marker='o', label='B')
plt.plot(Pc_x, Pc_y, marker='o', label='C')
plt.plot(Pd_x, Pd_y, marker='o', label='D')
plt.legend()
plt.show()
```
