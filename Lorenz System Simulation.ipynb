```python
Z1 = [112358]
U1 = []
rv = []
for i in range(64):
    x = (Z1[-1]**2) % 10**9
    Z1.append(int(x/10**3))
    U1.append(Z1[-1]/10**6)
    rv.append(U1[-1])
print("Initial",rv)

Z2 = [11,23,58]
a, c, M = 3, 7, 63
for i in range(16):
    Z2.append((a*Z2[-1] + Z2[-2] + c*Z2[-3]) % M)
    x = (Z1[-1]**2) % 10**9
    Z1.append(int(x/10**3))
    U1.append(Z1[-1]/10**6)
    rv[Z2[i+3]] = U1[-1]
print("Final  ",rv)

import matplotlib.pyplot as plt
plt.hist(rv)
plt.show()
```
