```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(99)

trails = 10000000
hits = 0
hit_x = []
hit_y = []
miss_x = []
miss_y = []

for i in range(trails):
    x = np.random.uniform(low = 0, high = 4)
    y = np.random.uniform(low = 0, high = 2)
    if x <= 2:
        if y <= x:
            hits += 1
            hit_x.append(x)
            hit_y.append(y)
        else:
            miss_x.append(x)
            miss_y.append(y)
    else:
        if (x -2)**2 + y**2 <= 2**2:
            hits += 1
            hit_x.append(x)
            hit_y.append(y)
        else:
            miss_x.append(x)
            miss_y.append(y)
area = (hits/trails)*8
print(area)
plt.scatter(hit_x,hit_y, c='red',s=0.1)
plt.scatter(miss_x, miss_y, c='blue',s=0.1)
plt.show()
```
