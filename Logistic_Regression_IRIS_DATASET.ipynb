```python
# load datasets using sklearn library
from sklearn import datasets
iris = datasets.load_diabetes()
features = iris.data[:, :2]
target = (iris.target != 0) * 1
```

```python
import numpy as np
import random
ft = list(zip(features, target)) # for same sequence shuffle in both list, we first need to combine it using zip
random.shuffle(ft)
f, t = zip(*ft) # unzipped value is assigning between two lists
fl = list(f) # tupple to list(not proper list)
tl = list(t) # tupple to list(not proper list)
fll = np.array(fl) # again reconverting list to an array
tll = np.array(tl) # again reconverting list to an array
X = fll.tolist() # properly converted to list from numpy array
y = tll.tolist() # properly converted to list from numpy array
for row in X:
    row.insert(0,1) # insert 1 in initial index
```

```python
# split dataset
Train_set_X=[]
Train_set_y=[]
Val_set_X=[]
Val_set_y=[]
Test_set_X=[]
Test_set_y=[]
random.seed(3)

for a, b in zip(X, y): # iterate two list at a time
    R = random.uniform(0,1)
    if R>=0 and R<=0.7:
        Train_set_X.append(a)
        Train_set_y.append(b)
    elif R>0.7 and R<=0.85:
        Test_set_X.append(a)
        Test_set_y.append(b)
    else:
        Val_set_X.append(a)
        Val_set_y.append(b)
```

```python
from random import seed
from random import random
import numpy as np
import math
l_r = input('Learning_Rate?: ')
def sigmoid(x): # sigmoid function
  return 1 / (1 + math.exp(-x))
train_loss = []
seed(1)
theta = [random(), random(), random()] # randomly initializing the theat's value

for i in range(0,1000):
    TJ = 0 # every time before entering into inner loop count of total cross-entropy value is intial with 0
    for m, n in zip(Train_set_X, Train_set_y):
        m = np.array(m) # convert into numpy array because python can't multiply a float multi-columed list using a single value
        n = np.array(n) # convert into numpy array because python can't multiply a float multi-columed list using a single value
        z = np.dot(m, theta) # dot product using numpy function
        h = sigmoid(z) # calling sigmoid funtion
        J = -n*np.log(h) -(1-n)*np.log(1-h) # log loss(cross-entropy loss) function
        TJ =TJ + J # continuous sum of cross-entropy loss
        dv = m*(h - n) # derivative
        theta = theta - dv*float(l_r) # updating theta's value
    ATJ = TJ/len(Train_set_X) # Average total loss
    train_loss.append(ATJ) 

import matplotlib.pyplot as plt
train_loss = np.array(train_loss)
e = np.arange(0,1000) # X_axis
g = train_loss # Y_axis
plt.plot(e,g)
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.title('Train_loss vs Epoch graph')
plt.show()


correct = 0
for m, n in zip(Val_set_X, Val_set_y): # iterate two list at a time
        m = np.array(m) # convert into numpy array because python can't multiply a float multi-columed list using a single value
        n = np.array(n) # convert into numpy array because python can't multiply a float multi-columed list using a single value
        z = np.dot(m, theta) # dot product using numpy function
        h = sigmoid(z) # calling sigmoid funtion
        if h>=.5: # ceil the values to 1
            h = 1
        else: # floor the values to 0
            h = 0
        if h==n: # counting the correct values
            correct = correct+1

Val_acc = (correct/len(Val_set_X))*100 # Validation dataset's accuracy
print(Val_acc,"%")
```
