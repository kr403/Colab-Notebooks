## Postfix Evaluation

```python
expression = input("Enter your postfix expression: ")
expression = expression.replace(" ", "")
operand = 0
signList = 0
count = 0
for ch in expression:
    if ch.isnumeric():
        operand += 1
    elif ch == '^' or ch == '*' or ch == '/' or ch == '+' or ch == '-':
        signList += 1
    elif ch == '(':
        count += 10
    elif ch == ')':
        count -= 10
count = operand - signList - 1
if count == 0:
    print("Valid Expression")
    digits = []
    sign = ['^', '*', '/', '+', '-']
    for ch in expression:
        if ch.isnumeric():
            digits.append(ch)
        else:
            y = int(digits.pop())
            x = int(digits.pop())
            if ch == '^':
                x = x ** y
                digits.append(x)
            elif ch == '*':
                x = x * y
                digits.append(x)
            elif ch == '/':
                x = x / y
                digits.append(x)
            elif ch == '+':
                x = x + y
                digits.append(x)
            elif ch == '-':
                x = x - y
                digits.append(x)
    print(digits[0])
else:
    print("Invalid Expression")
```
