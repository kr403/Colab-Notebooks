```
import copy

from typing import List
def Init():
  list = [2, 1, 5, 0, 8, 4, 10, 0, 20, 10]
  return list

def calc_cost(state):
  count = 0
  for element in range(0, len(state)):
    for i in range(element + 1, len(state)):
      if state[element] > state[i]:
        count += 1
        #print("Count =", count)
  #print("Total Count =", count)
  return count

def swap(list, a, b):
  list[a], list[b] = list[b], list[a]
  return list

def State_generation(current_state, current_state_cost):
  cost = current_state_cost
  state = current_state

  for element in range(0, len(current_state)-1):
    new_state = copy.copy(current_state)
    new_state = swap(new_state, element, element+1)
    print("New State:", new_state)

    state_cost = calc_cost(new_state)

    if(state_cost < current_state_cost):
      current_state_cost = state_cost
      #print("Current State Cost:", current_state_cost)
      state = new_state

  if(cost == state_cost):
    return current_state, None
  else:
    return state, state_cost

def goal_test(state):
  if calc_cost(state) == 0:
    return True
  else:
    return False

state = Init()

cost = calc_cost(state)
while(goal_test(state) == False):
  state, cost = State_generation(state, cost)

  if(cost == None):
    #print("State:", state)
    break
print("State:", state)
```
