```python
#Steepest Ascent

arr = [4,1,2,4,5]


def calcost(arr):
  cost = 0
  for i in range(len(arr)):
    for j in range(i,len(arr)):
      if arr[i] > arr[j]:
        cost=cost+1
  return cost

def steepStatgen(nt, cost):
  arr = nt.copy()
  while goaltest(arr)!=True:
    arr = nt.copy()
    for i in range(len(arr)):
      for j in range(i+1,len(arr)):
        arr[i], arr[j] = arr[j], arr[i]
        print("stategen -",arr,"Cost -",calcost(arr))
        if calcost(arr) >= cost :
          arr[i], arr[j] = arr[j], arr[i]
        elif goaltest(arr):
          return arr
        else:
          nt = arr.copy()
          cost = calcost(arr)
          arr[i], arr[j] = arr[j], arr[i]
    print("Next State -",nt,"Cost -",calcost(nt))
  return nt

def goaltest(arr):
  if calcost(arr) == 0:
    return True
  else :
    return False


print(steepStatgen(arr,calcost(arr)))
```
