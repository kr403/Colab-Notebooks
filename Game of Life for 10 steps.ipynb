## 1

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(765)

trails = 1000000
hits = 0
hit_x, hit_y, miss_x, miss_y = [], [], [], []

for i in range(trails):
    x = np.random.uniform(low = 2, high = 8)
    y = np.random.uniform(low = 0, high = 6)
    if x<=6 and (x-4)**2 + (y-4)**2<=2**2 or x<=6 and y<4:        
        hits += 1
        hit_x.append(x)
        hit_y.append(y)
    elif x>6 and y>=2 and (x-6)+(y-2)<=2 or x>6 and y<2 and (x-6)*y<=y**2:
        hits += 1
        hit_x.append(x)
        hit_y.append(y)
    else:
        miss_x.append(x)
        miss_y.append(y)

area = (hits/trails)*36
print(area)
plt.scatter(hit_x,hit_y, c='red', s = .1)
plt.scatter(miss_x, miss_y, c='blue', s = .1)
plt.show()
```

## 2

```python
import math
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(575)

xPoints, yPoints = [], []
trials = 100000
ll = 1
ul = 2
sumF = 0
sumSqF = 0

for i in range(trials):
    x = np.random.uniform(low=ll, high=ul)
    y = (((3 * x ) + 2) * np.log(x))
    sumF += y
    sumSqF += y**2
    xPoints.append(x)
    yPoints.append(y)

avgSumF = sumF/trials
avgSumSqF = sumSqF/trials
integral = avgSumF*(ul-ll)
error = (ul-ll)/np.sqrt(trials)*np.sqrt(avgSumSqF - avgSumF**2)
print(integral)
print(error)
plt.plot(xPoints, yPoints, c="gray")
plt.show()
```

```python
import numpy as np

matrix = np.zeros((12, 12), dtype=np.int32)
matrix[4:7,5],matrix[5,6] = 1,1
resultantMatrix = np.copy(matrix)
count = 0
print("Initial grid (t = 0)")
print(matrix)

for epoch in range(10):
    for ci,i in enumerate(matrix):
        for cj,j in enumerate(i):
            count = 0

            if ci-1>=0 and cj-1>=0 and matrix[ci-1,cj-1] == 1:    #i-1,j-1               
                count += 1
            if ci>=0 and cj-1>=0 and matrix[ci,cj-1] == 1:        #i,j-1                
                count += 1
            if ci+1<=11 and cj-1>=0 and matrix[ci+1,cj-1] == 1:   #i+1,j-1                
                count += 1
            if ci+1<=11 and cj>=0 and matrix[ci+1,cj] == 1:       #i+1,j                
                count += 1
            if ci+1<=11 and cj+1<=11 and matrix[ci+1,cj+1] == 1:  #i+1,j+1                
                count += 1
            if ci>=0 and cj+1<=11 and matrix[ci,cj+1] == 1:       #i,j+1                
                count += 1
            if ci-1>=0 and cj+1<=11 and matrix[ci-1,cj+1] == 1:   #i-1,j+1                
                count += 1
            if ci-1>=0 and cj>=0 and matrix[ci-1,cj] == 1:        #i-1,j
                count += 1

            if count >= 3 and j == 0:
                resultantMatrix[ci,cj] = 1
            elif count <= 1 and j == 1 or count >= 4 and j == 1:
                resultantMatrix[ci,cj] = 0

    matrix = np.copy(resultantMatrix) 
    print("t = ",epoch+1)
    print(resultantMatrix)
```
