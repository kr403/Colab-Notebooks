```python

import numpy as np
import pandas as pd
import random
import math

FILENAME = 'diabetes.csv'
lr = 0.00001

# 1. Load Data into numpy array
df = pd.read_csv(FILENAME)
data = df.to_numpy()



# 2. Split Dataset to Train/Validation/Test set according to the provided algorithm
training_set = []
validation_set = []
test_set = []
random.shuffle(data)

for row in data:
  rand = random.random()
  if 0<=rand<=0.8:
    training_set.append(row)
  elif 0.8<rand<=0.9:
    validation_set.append(row)
  else:
    test_set.append(row)


# 3. Implement the Logistic Regression Classifier Algorithm
def h(t):
  return 1.0 / (1 + np.exp(-t))

cols = int(data.shape[1])
thetas = np.zeros(cols) # initially all zeroes
m = len(training_set)


last_J = 99999
for e in range(200):
    preds = []
    for row in training_set:
        value = 0
        for i in range(cols):
            value += (row[i]*thetas[i])
        preds.append(h(value))

    L = 0
    i=0
    for row in training_set:
        y = row[-1]
        L += ( (-y*np.log(preds[i])) -((1-y)*np.log(1-preds[i])) )
        i+=1
    
    J = (1/m) * L

    if last_J-J<0.001:
      break
    last_J = J

    dThetas = np.zeros(cols-1)
    for i in range(len(training_set)):
      for j in range(cols-1):
        dThetas[j] += (training_set[i][j]*(preds[i]-training_set[i][-1]))
    

    for i in range(cols-1):
      dThetas[i] /= m

    for i in range(cols-1):
      thetas[i] -= (dThetas[i]*lr)

# Classification Accuracy on validation set
correct_outcome = 0
validation_pred = []
for row in validation_set:
  value = 0
  for i in range(cols):
      value += (row[i]*thetas[i])
  validation_pred.append(h(value))

i=0
for pred in validation_pred:
  outcome = 0
  if pred>=0.5:
    outcome=1
  if outcome == validation_set[i][-1]:
    correct_outcome+=1
  i+=1
print("Validation Accuracy - "+ str(correct_outcome/len(validation_set)))

# Classification Accuracy on test set
i=0
correct_outcome=0
test_pred = []
for row in test_set:
  value = 0
  for i in range(cols):
      value += (row[i]*thetas[i])
  test_pred.append(h(value))

i=0
for pred in test_pred:
  outcome = 0
  if pred>=0.5:
    outcome=1
  if outcome == test_set[i][-1]:
    correct_outcome+=1
  i+=1
print("Test Accuracy - "+ str(correct_outcome/len(test_set)))
```
