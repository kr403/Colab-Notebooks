## 1

```python
n = int(input("n = "))# Convert n into an intiger
s = input("s = ")
split_word_list = s.split()# Split the s string
if n != len(split_word_list):
    print("Word count mismatch. Try again!")
else:
    maximum_length = 0
    for word in split_word_list:
        if (len(word) > maximum_length):
            maximum_length = len(word)
    for i in range(n + 2):
        if (i == 0 or i == n + 1):
            
            # Print the first and last line's asterisk symbols
            print("*" * (maximum_length + 2))
        else:
            # Print every lines 1st asterisk symbol except first and last line
            print("*", end = "")
            print(split_word_list[i - 1], end = "")
            space_length = maximum_length - len(split_word_list[i - 1])
            # Print every lines space symbol except first and last line
            print(" " * (space_length), end = "")
            # Print every lines last asterisk symbol except first and last line
            print("*")
```

## 2

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(140)

trials = 1000
hits = 0

hit_x = []
hit_y = []

miss_x = []
miss_y = []

for i in range(trials):
    x = np.random.uniform(low=-8, high=8)
    y = np.random.uniform(low=-6, high=6)

    # Equation of a circle = ( x - h )^2 + ( y - k )^2 = r^2
    # Where ( h, k ) is the center and r is the radius.
    h, k, r = 3, 1, 5
    # Hit condition,
    if (x-h) ** 2 + (y-k) ** 2 <= r ** 2:
        hits += 1
        hit_x.append(x)
        hit_y.append(y)
    else:
        miss_x.append(x)
        miss_y.append(y)

print("Number of hits =", hits)
# Area of circle = πr^2 =25π
# Area of rectangle = wl [w = width = 12, l = length =16] = 192
# P(Points inside the circle) = 25π/192
# π = (192/25) * P(Points inside the circle)
est_pi = (192 / 25) * ((hits) / (trials))
print("Estimated π =", est_pi)
print("Real value of π =", np.pi)

plt.scatter(hit_x, hit_y, c='red')
plt.scatter(miss_x, miss_y, c='blue')
plt.show()
```
