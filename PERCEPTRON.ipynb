# Task 1(PerceptronN)

```python
import pandas as pd
import numpy as np
from numpy import random

omega = random.rand(9)

class PerceptronN:
    def __init__(self, maxIteration):
        self.maxIteration = maxIteration

    def fit(self, x_train, y_train):
        global omega
        missClassifiedIndex = []
        missClassifiedSign = []
        count = 1
        while True:
            missClassifiedIndex.clear()
            missClassifiedSign.clear()
            for index, xi in enumerate(x_train):
                y_dot = np.dot(xi, omega)
                if y_dot >= 0:
                    y_dot = 1
                else:
                    y_dot = 0
                if y_train[index] == 0 and y_dot == 1:
                    missClassifiedIndex.append(index)
                    missClassifiedSign.append(-1)
                elif y_train[index] == 1 and y_dot == 0:
                    missClassifiedIndex.append(index)
                    missClassifiedSign.append(1)
            for index, sign in zip(missClassifiedIndex, missClassifiedSign):
                if sign == -1:
                    omega = omega - x_train[index]
                elif sign == 1:
                    omega = omega + x_train[index]
            if count == self.maxIteration or len(missClassifiedIndex) == 0:
                break
            else:
                count += 1

    def predict(self, x_test):
        predictions = []
        for xi in x_test:
            y_dot = np.dot(xi, omega)
            if y_dot >= 0:
                y_dot = 1
                predictions.append(y_dot)
            else:
                y_dot = 0
                predictions.append(y_dot)
        return predictions
```

# Task 2(PerceptronRP)

```python
import pandas as pd
import numpy as np
from numpy import random

omega = random.rand(9)

class PerceptronRP:
    def __init__(self, maxIteration):
        self.maxIteration = maxIteration

    def fit(self, x_train, y_train):
        global omega
        count = 1
        while True:
            change = 0
            for index, xi in enumerate(x_train):
                y_dot = np.dot(xi, omega)
                if y_dot >= 0:
                    y_dot = 1
                else:
                    y_dot = 0
                if y_train[index] == 0 and y_dot == 1:
                    omega = omega - xi
                    change += 1

                elif y_train[index] == 1 and y_dot == 0:
                    omega = omega + xi
                    change += 1
            if count == self.maxIteration or change == 0:
                break
            else:
                count += 1 

    def predict(self, x_test):
        predictions = []
        for xi in x_test:
            y_dot = np.dot(xi, omega)
            if y_dot >= 0:
                y_dot = 1
                predictions.append(y_dot)
            else:
                y_dot = 0
                predictions.append(y_dot)
        return predictions
```
