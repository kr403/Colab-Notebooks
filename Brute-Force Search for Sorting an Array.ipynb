```python
def init():
    List = [2, 1, 5, 0, 8, 4, 10, 0, 20, 10]
    return List


def calc_cost(state):
    cost = 0
    for x in range(0, len(state)):
        for y in range(x + 1, len(state)):
            if state[x] > state[y]:
                cost += 1
    return cost


def state_generation(current_state, current_state_cost):
    cost = 999999
    for x in range(0, len(current_state)):
        for y in range(x + 1, len(current_state)):
            new_arr = []
            for i in current_state:
                new_arr.append(i)
            temp = new_arr[x]
            new_arr[x] = new_arr[y]
            new_arr[y] = temp
            new_cost = calc_cost(new_arr)
            if new_cost < cost:
                cost = new_cost
                return_state = []
                return_state = new_arr

    if current_state_cost > cost:
        return return_state, new_cost
    else:
        return current_state, None


def goal_test(state):
    if calc_cost(state) == 0:
        return True
    else:
        return False


if __name__ == '__main__':
    state = init()
    while not goal_test(state):
        state, cost = state_generation(state, calc_cost(state))
        if cost is None:
            print(f'Sorted Array: {state}')
            exit()
    print(f'Sorted Array: {state}')
```
