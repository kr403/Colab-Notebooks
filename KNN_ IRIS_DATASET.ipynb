```
# Dataset location in Google Drive
from numpy import genfromtxt
url = '/content/drive/MyDrive/Ai_Dataset/iris.csv'

# Dataset load as using NumPy package as NumPy array
iris = genfromtxt(url, delimiter =',')
print(type(iris))
```

```
# Convert NumPy array to Python list
datasets = iris.tolist()
for row in datasets:
  print(row)
```

```
# Shuffle datasets using random package
import random
random.shuffle(datasets)
print(datasets)
```

```
# Manually split dataset into three lists as Train(70%), Validation(15%) and Train(15%) set
Train_set=[]
Val_set=[]
Test_set=[]
for row in datasets:
  R = random.uniform(0,1)
  if R>=0 and R<=0.7:
    Train_set.append(row)
  elif R>0.7 and R<=0.85:
    Val_set.append(row)
  else:
    Test_set.append(row)
```

```
import numpy # To use NumPy package functionality
import math # To use some math functionality
import copy # To actually copy a list(to avoid reference pointing manipulation problems)
K = input('K?: ') # Take input for K
count = 0; # To count the total VALIDATION samples
right = 0; # To count the correct VALIDATION samples
L = [] # To store the Train sample list with Euclidean distance
for row in Val_set: # Loop executes until the Validation sample list ends
  V = row
  for row in Train_set: # Loop executes until the Training sample list ends
    T = row
    t = row.copy() # Copy the Training sample list in real
    distance = math.sqrt((V[0]-T[0])**2 + (V[1]-T[1])**2 + (V[2]-T[2])**2 + (V[3]-T[3])**2) # Euclidean distance formula
    t.append(distance)
    L.append(t)

  L.sort(key=lambda x: x[5]) # Sort the Train sample list with Euclidean distance in respect to based on Euclidean distance
  
  LK = L[0:int(K)] # Store first K^th Training sample
  print(LK)

  C = [] # To store the classification value of first K^th Training sample
  for i in range(0,int(K)):
    C.append(LK[i][4]) # Add the classification value of first K^th Training sample
  print(C)

  # Find the majority class
  idx, ctr = 0, 1
  for i in range(1, int(K)):
    if C[idx] == C[i]:
      ctr += 1
    else:
      ctr -= 1
      if ctr == 0:
        idx = i
        ctr = 1
  majority = C[idx]
  print(majority)

  # Check if majority class is right or wrong in respect to Validation sample class
  if(V[4] == majority):
    print("correct")
    right = right + 1 # Increase right value by one
  else:
    print("wrong")

  # Increse count value by one
  count = count + 1

print(right)
print(count)
accuracy = (right * 100)/count # Calculate accuracy in respect to the total validation sample
print(accuracy)
```
